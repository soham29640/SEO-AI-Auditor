```python
import json
import pandas as pd

path = "../data/crawl_results.json"

with open(path, "r", encoding="utf-8") as f:
    data = json.load(f)

df = pd.DataFrame(data)

df["clean_text"] = (
    df["headings"].apply(lambda x: " ".join(x) if isinstance(x, list) else "")
    + " "
    + df["content"].fillna("")
)

df.to_json(
    "../data/crawl_results_with_clean_text.json",
    orient="records",
    indent=2,
    force_ascii=False
)
```
